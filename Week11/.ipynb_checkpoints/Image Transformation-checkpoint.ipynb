```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
Image(filename='./Images/Honeycrisp.jpg')
```

```python
blue_apple = cv2.imread('./Images/Honeycrisp.jpg',1)
plt.imshow(blue_apple)
plt.figure()
red_apple = cv2.cvtColor(blue_apple, cv2.COLOR_BGR2RGB)
plt.imshow(red_apple)
```

```python
cropped_apple = red_apple[50:300, 400:550]
plt.imshow(cropped_apple)
```

```python
width = 50
aspect_ratio = width / cropped_apple.shape[1]
height = int(cropped_apple.shape[0] * aspect_ratio)
size = (width, height)
resized_leaf = cv2.resize(cropped_apple, size, interpolation=cv2.INTER_CUBIC)
resized_leaf2 = cv2.resize(cropped_apple, size, interpolation=cv2.INTER_AREA)
#resized_leaf = cv2.resize(cropped_apple,None,fx=2, fy=2)
#resized_leaf = cv2.resize(cropped_apple, tuple([7*x for x in size]), interpolation=cv2.INTER_CUBIC)
#resized_leaf2 = cv2.resize(cropped_apple, tuple([7*x for x in size]), interpolation=cv2.INTER_AREA)
plt.imshow(resized_leaf)
plt.title('Cubic')
plt.figure()
plt.title('Area')
plt.imshow(resized_leaf2)
```

```python
#cv2.imwrite("./Images/resized_leaf_big_cubic.jpg", resized_leaf)
#cv2.imwrite("./Images/resized_leaf_big_area.jpg", resized_leaf2)
#cv2.imwrite("./Images/resized_leaf_small_cubic.jpg", resized_leaf)
#cv2.imwrite("./Images/resized_leaf_small_area.jpg", resized_leaf2)
```

```python
leaf_flipped_vertical = cv2.flip(cropped_apple, 0)
leaf_flipped_horizontal = cv2.flip(cropped_apple, 1)
leaf_flipped_both = cv2.flip(cropped_apple, -1)
plt.figure(figsize=[10,3])
plt.subplot(141)
plt.imshow(cropped_apple)
plt.subplot(142)
plt.imshow(leaf_flipped_vertical)
plt.subplot(143)
plt.imshow(leaf_flipped_horizontal)
plt.subplot(144)
plt.imshow(leaf_flipped_both)
```

```python
img = cv2.imread('./Images/letter_on_table.jpg',1)
letter_on_table = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(letter_on_table.shape)
#flipping output image
cols = int(letter_on_table.shape[0]/2)
rows = int(letter_on_table.shape[1]/2)

plt.imshow(letter_on_table)
#top left
plt.plot(600,310, "og")
#top right
plt.plot(1050,290, "og")
#bottom left
plt.plot(595,860, "og")
#bottom right
plt.plot(1180,825, "og")
plt.show()
```

```python
pts1 = np.float32([[600,310],[1050,290],[595,860],[1180,825]])
pts2 = np.float32([[0,0],[450,0],[0,800],[450,800]])
M = cv2.getPerspectiveTransform(pts1,pts2)
letter_on_table_warped = cv2.warpPerspective(letter_on_table,M,(cols, rows))
plt.imshow(letter_on_table_warped)
```
